```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def metricas_test(modelo, X_test, y_test):
    pred = modelo.predict(X_test)
    return {
        "Accuracy":  accuracy_score(y_test, pred),
        "Precision": precision_score(y_test, pred, zero_division=0),
        "Recall":    recall_score(y_test, pred, zero_division=0),
        "F1":        f1_score(y_test, pred, zero_division=0),
    }

comp = pd.DataFrame({
    "LogisticRegression": metricas_test(pipe_log, X_test, y_test),
    "RandomForest":       metricas_test(rf,       X_test, y_test)
}).T.sort_values("F1", ascending=False)

print("=== Comparación de modelos (SET DE PRUEBA) ===")
display(comp)
```

```python
def diagnostico_over_under(modelo, nombre, X_train, y_train, X_test, y_test):
    from sklearn.metrics import f1_score
    ytr = modelo.predict(X_train)
    yte = modelo.predict(X_test)
    f1_tr = f1_score(y_train, ytr, zero_division=0)
    f1_te = f1_score(y_test,  yte, zero_division=0)
    gap = f1_tr - f1_te
    print(f"== {nombre} ==")
    print(f"F1 Train: {f1_tr:.3f} | F1 Test: {f1_te:.3f} | Gap: {gap:.3f}")
    if gap > 0.10 and f1_te < f1_tr:
        print("Diagnóstico: OVERFITTING (aprende demasiado el train, generaliza peor).")
    elif f1_tr < 0.65 and f1_te < 0.65:  # umbral orientativo, ajusta si quieres
        print("Diagnóstico: UNDERFITTING (el modelo es demasiado simple o falta señal).")
    else:
        print("Diagnóstico: Balance razonable (sin señales fuertes de over/underfitting).")
    print("-"*50)

diagnostico_over_under(pipe_log, "LogisticRegression", X_train_bal, y_train_bal, X_test, y_test)
diagnostico_over_under(rf,       "RandomForest",       X_train_bal, y_train_bal, X_test, y_test)
```

# Conclusión y Recomendaciones (Telecom X – Parte 2)

## Hallazgos clave
- Variables más relevantes para Churn:  
  - `antiguedad_meses`: relación negativa → los clientes más nuevos tienden a cancelar.  
  - `cargos_totales`: relación negativa → clientes con mayor gasto acumulado permanecen más.  
  - `cargos_mensuales` y `facturacion_online`: correlación positiva con la cancelación.  

## Desempeño de modelos (set de prueba)
- Regresión Logística → F1 ≈ 0.58  
- Random Forest → F1 ≈ 0.59  
- Ambos presentan overfitting: muy buen rendimiento en entrenamiento, menor en prueba.  

## Análisis crítico
- Random Forest tuvo un F1 levemente superior, pero con mayor sobreajuste.  
- Regresión Logística es más interpretable y permite comunicar mejor los factores de riesgo.  
- Posibles causas del sobreajuste:  
  - Uso de SMOTE (datos sintéticos).  
  - Complejidad alta en Random Forest.  
  - Hiperparámetros sin optimizar.  

## Recomendaciones de negocio
1. Enfocar estrategias de retención en clientes nuevos (primeros meses de contrato).  
2. Revisar planes con cargos mensuales altos, ofreciendo descuentos o bundles.  
3. Incentivar la migración de contratos mensuales a anuales con beneficios.  
4. Mejorar la experiencia de facturación online para reducir frustración y bajas.  
5. Implementar alertas tempranas para clientes con baja antigüedad y altas cuotas mensuales.  

## Próximos pasos técnicos
- Regularización en la logística (`C`) y ajuste de hiperparámetros en Random Forest.  
- Usar validación cruzada estratificada con SMOTE dentro de un pipeline.  
- Optimizar el umbral de decisión para mejorar recall y F1.  
- Evaluar métricas costo-sensibles, ya que un falso negativo implica pérdida de cliente.  
- Utilizar métodos interpretativos (coeficientes, importancias, SHAP) para explicar decisiones.  

---

**Resumen**: La antigüedad baja y los patrones de cobro/contrato son los principales predictores de churn.
